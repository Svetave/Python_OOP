So, you must define the operations for addition, multiplication, subtraction, division and absolute value of complex numbers. Also you need to define the conjugate method.

Note: Your program should not read or print anything.

```python
class CompleX:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y  
    def __add__(self, other):
        if isinstance(other, int):
            x = other 
            other = CompleX(x, 0)
            return self.x+other.x, self.y+other.y
        else:
            return self.x+other.x, self.y+other.y       
    def __sub__(self, other):
        return self.x-other.x, self.y-other.y              
    def __mul__(self, other):
        if isinstance(other, int):
            x = other 
            other = CompleX(x, 0)
            return self.x*other.x-self.y*other.y, self.x*other.y+self.y*other.x
        else:
            return self.x * other.x - self.y * other.y, self.x * other.y + self.y * other.x       
    def __truediv__(self, other):
        return (self.x*other.x+self.y*other.y)/(other.x**2+other.y**2), (other.x*self.y-self.x*other.y)/(other.x**2+other.y**2)   
    def conjugate(self):
        return self.x, -self.y  
    def __str__(self):
        return '({self.x}, {self.y})'.format(self=self)
    def __abs__(self):
        absolute = (self.x**2+self.y**2)**0.5
        return absolute
```

2. Implement class Translator. Describe the following methods inside the class:
* **fill_dictionary(self, inp_dict)** -- takes a dictionary as argument and writes word matching in the self.dict_ attribute
* **translate(self, text)** -- takes text as a string and translates text by using self.dict_. If some word is missing in the dictionary translate it as "unknown". The method should return a string with a translation.

Note: Your program should not read or print anything.

```python
#example
>>> inp_dict = {"привет": "hello", "пока": "bye", "кот": "cat"}
>>> ru_en = Translator()
>>> ru_en.fill_dictionary(inp_dict)
>>> text = "привет пока кот собака"
>>> res = ru_en.translate(text)
>>> print(res)
hello bye cat unknown
```

```python
class Translator:
    def fill_dictionary(self, inp_dict):
         self.dict_inp_dict=inp_dict
    def translate(self, text):
        res=''
        arr=text.split()
        for i in range (len(arr)):
            if arr[i] in self.dict_inp_dict:
                res+=' '
                res+=self.dict_inp_dict.get(arr[i])
            else:
                res+=' unknown'
        return res
```
