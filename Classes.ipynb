1. Implement the Book class. Your class should contain the following methods:

__init__*(self, book_name, author, number_of_pages)* -- creates an instance of a class with the book title book_name and author author and the number of pages number_of_pages.
* **get_title(self)** -- returns the book title
* **get_author(self)** -- returns the author of the book
* **get_pages(self)** -- returns the number of pages of a book
Note: The program should not read or output anything.

```python
class Book:
    def __init__(self, book_name, author, number_of_pages):
        self.book_name = book_name
        self.author = author
        self.number_of_pages = number_of_pages
    def get_title(self):
        return self.book_name
    def get_author(self):
        return self.author
    def get_pages(self):
        return self.number_of_pages
```

2. Implement the Fraction class. Your class should contain the following methods:

__init__*(self, a, b)* -- creates an instance of a class with numerator a and denominator b. It is guaranteed that a < b and b != 0. If possible, reduce the numerator and denominator by a common factor.
* **get_numerator(self)** -- returns the numerator
get_denominator(self) -- returns the denominator
* **__str__(self)**-- implement the method so that print returns the result as follows: (a / b)


```python
class Fractional:
    def __init__(self, a, b):
        cou = a
        while 1 != cou:
            if b%cou == 0 and a%cou == 0:
                b = b/cou
                a = a/cou
            cou -=1
        self.a = int(a)
        self.b = int(b)
    def __str__(self):
        return("(" + str(self.a)+" / "+str(self.b)+")")
    def get_numerator(self):
        return str(self.a)
    def get_denominator(self):
        return str(self.b)

```

3.Implement the Polynomial class. Your class should contain the following methods:

__init__(self, n) -- creates an instance of a class corresponding to a polynomial of degree n.
* **fill(self, args)** -- fills the first coefficients of the polynomial from the args list.
* **__str__(self)** -- to print the polynomial in the form as indicated in the example. Round up to one decimal places the values of coefficients in the output.



```python
class Polynomial:
    def __init__(self, n):
        self.coefficients = [0] * (n + 1)
    def fill(self, args):
        for i in range(len(args)):
            self.coefficients[i] = args[i]
    def __str__(self):
        a = str(round(float(self.coefficients[0]), 1))
        
        result = [a]
        for i in range(1, len(self.coefficients)):
            cv = round(float(self.coefficients[i]), 1) # cv -- current value
            cs = "+" if cv >= 0 else "-" # cs -- current symbol
            result.append(f"{cs} {abs(cv)} * x^{i}")
        return " ".join(result)

```

4.Implement the TypePrinter class. Your class should contain the following methods:

* __init__(self, obj) -- creates an instance of a class by some object
* **__str__(self)**  -- returns the class name of the object obj

```python
class TypePrinter:
    def __init__(self, obj):
        self.obj = obj
    def __str__(self):
        return f'{type(self.obj)}'
```
