1. Implement the Text class with the following methods:

__init__(self, text) - object initialization with some text text

* get_number_of_lines(self) - returns the number of lines

* get_line(self, index) - returns the index-th string of text

* get_nwords(self, index) - returns the number of words in the index-th line

* get_word(self, line_index, word_index) - returns word at indices (line_index, word_index)

Implement the child class EditableText with the following methods:

* change_line(self, new_line, index) - changes the line with index index to a new one

* change_word(self, new_word, line_index, word_index) - replaces the word at position (line_index, word_index) with a new one. Note: method does not change the punctuation mark if the word ends with it.

* get_substring_indxs(self, substr) - returns a list of positions of occurrence of a substring with overlapping in the text as list in the form [line index, word index inside a line]

The following example can be used to test your code:

```python
text = "aa bc aaaa\naab aaa, aaa"
etext_obj = EditableText(text)

print(etext_obj.get_number_of_lines()) # print 2

print(etext_obj.get_line(0)) # print aa bc aaaa

print(etext_obj.get_nwords(0)) # print 3

etext_obj.change_word("word", 1, 1)
etext_obj.change_line("aa aa aaaa", 0)

print(etext_obj.get_substring_indxs("aa")) # print
# [[0, 0], [0, 3], [0, 6], [0, 7], [0, 8], [1, 0], [1, 10], [1, 11]]

print(etext_obj.get_word(1, 1)) # print word,
```

```python
import re
class Text:
    def __init__(self, text):
        self.text = text
        
    def get_number_of_lines(self):
        return len(self.text.splitlines())
   
    def get_line(self, index):
        lines=self.text.splitlines()
        return lines[index]
            
    def get_nwords(self, index):
        lines=self.text.splitlines()
        line = lines[index]
        return len(line.split())
        
    def get_word(self, line_index, word_index):
        lines=self.text.splitlines()
        line = lines[line_index]
        line = line.split()
        return line[word_index]

class EditableText(Text):
    
    def change_line(self, new_line, index):
        lines = self.text.splitlines()
        lines[index]=new_line
        self.text="\n".join(lines)
     
    def change_word(self, new_word, line_index, word_index):
        lines=self.text.splitlines()
        line = lines[line_index]
        line = line.split()
        if ',' not in line[word_index]:
            line[word_index] = new_word
            line = ' '.join(line)
            lines[line_index] = line
            self.text="\n".join(lines)

        else:
            line[word_index] = new_word + ','
            line = ' '.join(line)
            lines[line_index] = line
            self.text="\n".join(lines)

    
    def get_substring_indxs(self, substr):
        lines=self.text.splitlines()
        a=[]
        r=[]
        for string in lines:
            a.append([el.start() for el in re.finditer('(?={0})'.format(re.escape(substr)), string)])
        for lst in a:
            r.append([[a.index(lst), el] for el in lst])
        flat = [x for sublist in r for x in sublist]
        return flat
```

Matrix:

RowsMetaData:

* def __init__(self, rownames) - creates an attribute for a list of row names
*  def get_rownames(self) - returns rownames as a list of strings

ColsMetaData: similar to previous

Eset:

* def __init__(self, vals_2d_list, rownames, colnames) - initializes an object by 1) a two-dimensional matrix of real numbers,  2) row names, 3) column names.
* def sort_cols(self, descending=False) - method sorts the matrix and column names by a list of column names in lexicographic order. If descending=True then sort in descending order
* sort_rows(self, descending = False) - similar to the previous one, only for matrix rows
* __str__(self) - returns a table with row and column names. Each value of the matrix should be displayed with 1 decimal places, and the number must be right-aligned with spaces to a length of 4. It is guaranteed that all values are in the range from 0 to 9
An example of how the code should work:

```python
eset_obj = Eset([[1, 2, 3], [4, 5, 6], [7, 8, 9]], ["103", "102", "10"], ["A3", "A2", "A1"])

print(eset_obj) # result should be:
#        A3   A2   A1
#  103  1.0  2.0  3.0
#  102  4.0  5.0  6.0
#   10  7.0  8.0  9.0

eset_obj.sort_cols()
print(eset_obj) # result should be:
#        A1   A2   A3
#  103  3.0  2.0  1.0
#  102  6.0  5.0  4.0
#   10  9.0  8.0  7.0


eset_obj.sort_rows(descending=True) # result should be:
print(eset_obj) # result should be:
#        A1   A2   A3
#  103  3.0  2.0  1.0
#  102  6.0  5.0  4.0
#   10  9.0  8.0  7.0


print("rownames: ", eset_obj.get_rownames()) # print: "rownames:  ['103', '102', '10']"
print("colnames: ", eset_obj.get_colnames()) # print: "colnames:  ['A1', 'A2', 'A3']"
```

```python
class Matrix:
    def __init__(self, vals_2d_list):
        self.vals_2d_list = vals_2d_list


class RowsMetaData:
    def __init__(self, rownames):
        self.rownames = rownames

    def get_rownames(self):
        return self.rownames


class ColsMetaData:
    def __init__(self, colnames):
        self.colnames = colnames

    def get_colnames(self):
        return self.colnames

class Eset(Matrix, RowsMetaData, ColsMetaData):
    def __init__(self, vals_2d_list, rownames, colnames):
        self.vals_2d_list = vals_2d_list
        self.rownames = rownames
        self.colnames = colnames

    def sort_cols(self, descending=False):
        for i in range(len(self.vals_2d_list)):
            for j in range(len(self.vals_2d_list[i])):
                self.vals_2d_list[i][j] = [self.colnames[j], self.vals_2d_list[i][j]]
        if not descending:
            for i in range(len(self.vals_2d_list)):
                self.vals_2d_list[i].sort()
            self.colnames.sort()
        elif descending:
            for i in range(len(self.vals_2d_list)):
                self.vals_2d_list[i].sort(reverse=True)
            self.colnames.sort(reverse=True)
        for i in range(len(self.vals_2d_list)):
            for j in range(len(self.vals_2d_list[i])):
                self.vals_2d_list[i][j] = self.vals_2d_list[i][j][1]


    def sort_rows(self, descending=False):
        for i in range(len(self.rownames)):
            self.vals_2d_list[i].insert(0, self.rownames[i])
        if not descending:
            self.vals_2d_list.sort()
            self.rownames.sort()
        elif descending:
            self.vals_2d_list.sort(reverse=True)
            self.rownames.sort(reverse=True)
        for i in range(len(self.vals_2d_list)):
            self.vals_2d_list[i].pop(0)

    def __str__(self):
        line = ['    ']
        result = ''
        for elem in self.colnames:
            line.append(f"{elem:>4}")
        line = " ".join(map(str, line))
        line += "\n"
        result += line
        for i in range(len(self.vals_2d_list)):
            line = [f"{self.rownames[i]:>4}"]
            for j in range(len(self.vals_2d_list[i])):
                elem = "%.1f" % self.vals_2d_list[i][j]
                line.append(f"{elem:>4}")
            line = " ".join(map(str, line))
            if i == len(self.vals_2d_list) - 1:
                result += line
            else:
                line += "\n"
                result += line
        return result
```
